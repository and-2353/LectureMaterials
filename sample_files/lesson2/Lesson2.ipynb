## 比較演算子
比較演算子を使うことで、条件式を作ることができます。真偽の判定ができます。

条件式の部分が最初に計算され、TrueもしくはFalseが出力されます。

以下のような比較演算子があります。
- ```==```
  - ```=```は代入なので、右辺と左辺が等しいか判定したいときは```==```を使います。

- ```<```や```<=```
  - わかりやすいと思う。

- ```!=```
  - ```!=```は、両辺が等しくない時Trueを返します。

````python
x = 1
print(x==1)
print(x<2)
print(x>2)
print(x<=2)  # ≦
print(1>=x)  # ≧
print("こん" != "ハロー")  # ≠
print(1 < 2 < 3)
````

## 条件分岐(if文)

条件分岐の書き方です。
```
if 条件式:
  [条件式がTrueの時の処理]
  :
  :
[Trueの処理終了後/Falseの場合、この行以降を処理する]
```

ifの後にはコロン```:```を置いて、次の行からインデント（字下げ）をします。

インデントは空白スペース4つ、空白スペース2つ、Tabスペース1つなどがあります。

インデントしている行がTrueの時に入る処理です。処理終了後はインデントされていない次の行に移ります。

Falseの時、インデントしている行には入らず、インデントされていない次の行に移ります。

if文とセットでelifとelseを覚えましょう。

## 繰り返し処理（for文）

for文には書き方が2パターンあります。**(本当は1つしかありません)**
1. 普通のループ
```
for 変数 in range(int):
    文
    ...
その後の文
```
変数に0が代入され、インデントされた行が実行されます。
インデントされた行の最後まで行ったら、forの行に戻ってきます。
これを```int```回繰り返します。

2. 配列```(list)```を使ったループ
```
for 変数 in list:
    文
    ...
その後の文
```
変数にlistの要素が順に代入され、インデントされた行が実行されます。
代入される値以外は同じ。





````python
for i in range(5):
  print(i)

print()

for fruit in ['apple', 'banana', 'grape']:
  print(fruit)
````

### range関数

range(5)≒[0, 1, 2, 3, 4]です。

厳密には左辺はrange型、右辺はlist型です。



## もう一度、for文
なので、for文の書き方1のパターンは、結局リスト（のようなもの）の要素が一個ずつ変数に代入されているにすぎないことが分かります。つまり、本質的には書き方は一つであり、for文の書き方を一般化すると以下のようになります。

```
for 変数 in iterable:
    文
    ...
その後の文
```

iterableとは順番に取り出せるもので，その例がlistやrangeです。他にもtupleがあります。tupleでfor文を使う例は珍しいと思う。
